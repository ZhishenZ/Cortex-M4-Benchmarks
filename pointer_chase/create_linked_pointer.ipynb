# Create 1000 linked list from l1 to l1000

```python
n = 1000
output_str = "node l"+str(n)+" = {NULL};"
for i in range(n):
    if((i%50)==0):
        output_str += "\n" 
    output_str +="node l"+str(n -1-i)+" ={&l" + str(n -i)+"};"
output_str += "node* n = &l0;"   
print(output_str)
```
