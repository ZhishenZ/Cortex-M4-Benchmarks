```python
import csv
data_list = []
with open('input_small.dat', 'r') as fd:
    reader = csv.reader(fd)
    for row in reader:
        data_list.append(row[0])
```

```python
output_string = ""
n = 50
for i in range(n-1):
    if i%50 ==0:
        output_string += "\n"
    output_string += "\""+ data_list[i] +  "\", "

output_string += "\""+data_list[n-1] + "\""
print(output_string)
```
